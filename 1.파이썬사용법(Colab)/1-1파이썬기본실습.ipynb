# 컴퓨터의 장점!!! 반복시키기

컴퓨터는 동일한 작업의 반복에 아주 강점이 있습니다!

반복시킬 수 있는 몇 가지 방법이 있지만 가장 많이 사용할 for문을 예로 들어 보겠습니다.

특정 횟수를 반복할 때는 다음과 같이 사용하면 됩니다.



```
for i in range(원하는 횟수):           # 콜론 필수! 문법입니다.
    print(i)                        # 들여쓰기 필수! 이것도 문법! 맞지 않을 경우 탭으로 조절하시면 됩니다.
```

위의 명령어 부분만 그리고 원하는 횟수에 자신이 원하는 숫자를 입력하고 아래 코드창에 실행시켜 보세요.


## for 문을 이용하여 피보나치수열을 찾아보겠습니다.



````python
a = 1              # 첫 번째 항은 1
print(a)          # 첫 번째 항 출력
b = 1              # 두 번째 항도 1
print(b)          # 두 번째 항 출력
for i in range(100):       # 50항까지 출력하기 위해 48번 더 반복하여 구함.
    c = a + b             # 앞의 두 항을 더해 c로 만듦. 즉 새롭게 만들어진 항
    print(c)
    a = b                 # 한 항씩 앞당겨 저장
    b = c
````

# 피보나치 수열을 리스트에 담아보기

````python
fibo_list = [1, 1]
for i in range(1000):               # 피보나치 수열의 1000번째 항까지 담긴 바구니 만들기
    fibo_list.append(fibo_list[i]+fibo_list[i+1])
fibo_list
````

원하는 항을 찾아 보기 위해서는



```
fibo_list[원하는 숫자]      # 단, 0이 첫 번째 항이므로 원하는 항보다 1작은 수 입력
```



````python
fibo_list[683]
````
